```python
import pandas as pd

meteo = pd.read_csv('export (3).csv')
meteo = meteo.drop(columns=['tsun'])

meteo
```

```python
import pandas as pd

# Suponemos que meteo ya tiene columnas: month, day, (sin 'date')
# Crear un rango de horas
horas = pd.DataFrame({'hour': list(range(24))})

# Repetir cada fila de meteo para cada hora (join cruzado)
meteo_expandida = meteo.merge(horas, how='cross')

# Mostrar estructura resultante
print(meteo_expandida.head(30))
```

```python
# Asegúrate de que la columna 'date' esté en formato datetime
meteo['date'] = pd.to_datetime(meteo['date'])

# Extraer año, mes, día y hora
meteo['year'] = meteo['date'].dt.year
meteo['month'] = meteo['date'].dt.month
meteo['day'] = meteo['date'].dt.day
meteo['hour'] = meteo['date'].dt.hour

# Eliminar la columna original 'date'
meteo = meteo.drop(columns=['date'])

# Comprobar que no hay nulos en las nuevas columnas
print(meteo[['month', 'day', 'hour']].isnull().sum())

# Mostrar las primeras filas
print(meteo.head())
```

```python
meteo['snow'] = meteo['snow'].notna().astype(int)
meteo['prcp'] = meteo['prcp'].notna().astype(int)
```

```python
import numpy as np

# Calcular viento medio como combinación de velocidad media y ráfaga máxima
meteo['wind'] = meteo[['wspd', 'wpgt']].mean(axis=1)

# Crear una columna categórica si el viento es fuerte (> 30 km/h por ejemplo)
meteo['wind_flag'] = np.where(meteo['wind'] > 30, '1', '0')
```

```python
meteo = meteo.drop(columns=['tmin', 'tmax', 'pres', 'wdir',	'wspd',	'wpgt'])
```
